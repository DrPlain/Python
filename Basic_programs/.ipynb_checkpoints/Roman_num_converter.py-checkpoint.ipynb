```python
def roman_to_num_converter(roman_numeral=""):
    roman_set = ('I', 'V','X','L','C','D','M')
    rom = roman_numeral
    if not isinstance(roman_numeral, str):
        return "Invalid input"
    if rom == "":
        return "Empty string"
    rom_set = set(rom)
    
    if rom_set.issubset(roman_set) == False:
        return "Invalid input"
```

```python
test_dict = {
    0:'Invalid input',
    "1":'Empty string',
    'LM':'Invalid input',
    'ABC':'Invalid input',
    'IIII':'Invalid input',
    'i':1,
    'I':1,
    'V':5,
    'IX':9,
    'X':10,
    'XL':40,
    'L':50,
    'XC':90,
    'C':100,
    'cL':140,
    'CDLVII':457,
    'DCCC':800,
    'CM':900,
    'MCDXCIV':1494,
    'MMMCDLVII':3457
}
```

```python
def test(test_dict):
    for key in test_dict:
        output = roman_to_num_converter(key)
        passed = 0
        if output == test_dict[key]:
            print("Roman numeral: {}\nYour output: {}\nExpected output: {}\nRemark: PASSES".format(key, output, test_dict[key]))
            print()
            passed += 1
        else:
            print("Roman numeral: {}\nYour output: {}\nExpected output: {}\nRemark: FAILED".format(key, output, test_dict[key]))
            print()
        
        print("Overall summary: {} out of {} tests passed".format(passed, len(test_dict)))
    return
```

```python
test(test_dict)
```
